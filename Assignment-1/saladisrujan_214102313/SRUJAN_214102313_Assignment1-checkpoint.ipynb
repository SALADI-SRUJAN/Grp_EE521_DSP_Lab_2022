**Plotting basic signals**

```python
def u(t):
    "unit step signal-->u(t)"
    u=[]
    for i in range(0,len(t)):
        if (t[i]>=0 and t[i]<=20):
            u.append(1)
        else:
            u.append(0)
    return u
def r(t):
    "ramp signal-->r(t)"
    r=[]
    for i in range(0,len(t)):
        if (t[i]>=0 and t[i]<=20):
            r.append(t[i]*u(t)[i])
        else:
            r.append(0)
    return r
    
    #return t*(1*(t >= 0))
def p(t):
    "parabolic signal-->p(t)"
    p=[]
    for i in range(0,len(t)):
        if (t[i]>=0 and t[i]<=20):
            p.append((t[i])**2*u(t)[i]/2)
        else:
            p.append(0)
    return p

functions = [u,r]
t = np.linspace(-21,21,100)
fig = plt.figure(10,figsize=(14,8))
for i,function in enumerate(functions,start=1):
    plt.subplot(2,2,i)
    plt.plot(t,function(t),'blue')
    plt.axhline(0,color='black',linewidth = 0.5)
    plt.axvline(0,color='black',linewidth = 0.5)
    plt.ylim((-0.30,20))
    plt.title(function.__doc__,color='red',size=20)
plt.subplot(2,2,3)
plt.plot(t,p(t),'blue')
plt.ylim(-0.30,200)
plt.title('parabolic signal-->p(t)',fontsize=20,color='red')
plt.tight_layout()
plt.show()
```

**(1)Convolution of two discrete signals**

```python
fig = plt.figure(1,figsize=(10,7))
#defining funtin for the convolution
def convolution(inputseq,impulseresponse):
        lengthA=np.size(inputseq)
        lengthB=np.size(impulseresponse)
        C = np.zeros(lengthA + lengthB -1)
        for m in np.arange(lengthA):
            for n in np.arange(lengthB):
                C[m+n] = C[m+n] + inputseq[m]*impulseresponse[n]
        return C 
a1=[3,5,2,6,8] #input sequences
b1=[-2,4,1]
n = np.arange(-10,10,1)
p=convolution(a1,b1) #performing convolution
print("output sequence=",p) #printing the convolution result
plt.subplot(3,1,1)
plt.stem(p,use_line_collection=True) #plotting convolution result.
plt.title('output',color='red',size=20)
plt.subplot(3,1,2)
plt.stem(a1,use_line_collection=True)
plt.title('input',color='red',size=20)
plt.subplot(3,1,3)
plt.stem(b1,use_line_collection=True)
plt.title('impulse sequence',color='red',size=20)
plt.tight_layout()
plt.show()
```
