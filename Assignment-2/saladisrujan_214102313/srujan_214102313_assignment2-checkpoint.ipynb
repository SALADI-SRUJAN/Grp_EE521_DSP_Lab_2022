**1.Probability of reconstruction of message without and with hamming codes**

```python
def nCr(n, r):#function to calculate combination ncr
 
    return (fact(n) / (fact(r)
                * fact(n - r)))

p=nCr(4,1)*((0.1)**1)*(0.9)**3+nCr(4,2)*((0.1)**2)*(0.9)**2+nCr(4,3)*((0.1)**3)*(0.9)**1+nCr(4,4)*((0.1)**4)*(0.9)**0 #without hamming code
print(f'Reconstruction prob without hamming code :{1-p}')
p_h=nCr(7,2)*((0.1)**2)*(0.9)**5+nCr(7,3)*((0.1)**3)*(0.9)**4+nCr(7,4)*((0.1)**4)*(0.9)**3+nCr(7,5)*((0.1)**5)*(0.9)**2+nCr(7,6)*((0.1)**6)*(0.9)**1+nCr(7,7)*((0.1)**7)*(0.9)**0
#with hamming code
print(f'Reconstruction prob with hamming code:{1-p_h}')
```

```python
print(f"Increase in reliability with hamming code:{((p-p_h)/(1-p))*100}%")
```

**3.Auto correlation of given x(n) sequence**

```python
def x(n):
    x=[]
    for i in n:
        if(i>=0 and i<=7):
            x.append(1)
        else:
            x.append(0)
    return x
n=np.arange(-10,11,1)
plt.subplot(3,1,1)
plt.stem(n,x(n))
plt.title('x(n)')
plt.subplot(3,1,2)
plt.stem(n,x(-n))
plt.title('x(-n)')
plt.tight_layout()
```

```python
def convolution(inputseq,impulseresponse):
        lengthA=np.size(inputseq)
        lengthB=np.size(impulseresponse)
        C = np.zeros(lengthA + lengthB -1)
        for m in np.arange(lengthA):
            for n in np.arange(lengthB):
                C[m+n] = C[m+n] + inputseq[m]*impulseresponse[n]
        return C 
```

```python
r=(convolution(x(n),x(-n)))##convolution of x(n) with x(-n) gives the auto correlation of x(n).
n1=np.arange(-20,21,1)
plt.stem(n1,r/8)
plt.title('Autocorrelation of x(n)')
plt.tight_layout()
plt.show()
```

```python
def x1(n):
    x1=[]
    for i in range(n):
        x1.append(np.random.normal(0,1))
    return x1
X1=x1(1000)
fig = plt.figure(10,figsize=(5,5))

plt.hist(X1)
plt.title('Histogram of 1000 samples')
plt.show()

      
```

```python
def r1(k,n):
    r1=0
    for i in range(n):
        if(i-k>=0):
            r1+=X1[i]*X1[i-k]
        else:
            r1+=0
    return r1/999

def R1(m,n):
    R1=[]
    for i in range(m):
        
        R1.append(r1(i,n))
    return R1
R_1=R1(100,1000)
R_2=R_1.copy()
def Reverse(lst):
    lst.reverse()
    return lst
plt.stem(np.arange(-100,100),Reverse(R_1)+R_2)
plt.title('Autocorrelation for m(100) lags of 1000 lags')
plt.show()
      
```

**Observation:The autocorrelation of the white gaussian is a impulse function at zero lag**

**10.a.Central chi-square distribution.**

```python
def chisquare_pdf(k): #k=degrees of freedom
    x=np.linspace(-10,10)
    y=np.piecewise(x,[(x>=0)&(x<=10)],[lambda x:(x**((k/2)-1)*(2.718**(-x/2)))/(2**(k/2)*math.gamma(k/2))])#pdf formula of chi square
    plt.plot(x,y,label=k)
    plt.title('pdf of chi-suare distribution')
    plt.xlabel('x')
    plt.ylabel('p(x)')
    plt.legend()
    plt.show()




```

**Non-central chisquared distribution**

```python
def central_chi_sqr(n):
    y_chi = []
    x_chi = np.linspace(0.01,10,1000)
    for i in x_chi:
        num = np.power(i,(n/2-1)) * np.exp(-i/2)
        den = np.power(2,n/2) * math.gamma(n/2)
        y_chi.append(num/den)
    return y_chi
```

```python
def non_central_chi_square(n,lam):
    y_non_cen_chi = 0
    var1 = 0
    var2 = 0
    
    for k in range(0,101):
        var1 = np.array(central_chi_sqr(n+2*k))
        var2 = np.exp(-lam/2) * np.power(lam/2,k) / math.factorial(k)
        y_non_cen_chi = y_non_cen_chi + var1*var2
    
    return y_non_cen_chi
#plotting non_central_chi_square(n,lambda) for different n,lambda
x=np.linspace(0.01,10,1000)

```

```python
plt.plot(x,  non_central_chi_square(2,1),  label='$k=2,\lambda=1$')
plt.title('Non Central Chi-Squared Distribution')
plt.legend()
plt.show()
```
